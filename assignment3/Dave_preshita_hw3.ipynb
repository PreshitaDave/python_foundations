# Part 1

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 19 13:11:40 2021

@author: preshitadave
"""

import matplotlib.pyplot as plt
import numpy as np
import random

class base_walker: 
    def __init__(self):
        pass
    def one_step(self):
        pass
    def steps(self,n=1):
        for i in range(n):
            self.one_step()
            
# inherit from base_walker to make a random_walker_equal class that:
# stores a sequence of integer positions, and,
# increments its position by +1, 0 or -1 every step

class random_walker_equal(base_walker):
    def __init__(self):
        # define stored variables, initial data
        self.x = 0
        self.walk = [self.x]
    def one_step(self):
        # increment the current position by +1 (right direction), 0 (stop) or -1 (left direction), with equal probability
        rand = random.random() #generates a random number in [0,1)
        if rand < 1/3:
            self.x -= 1 # moves to the left 
        elif rand >= 1/3 and rand < 2/3:
            self.x += 0 # stays at the same position
        else:
            self.x += 1 #moves to the right
        self.walk.append(self.x)
    def get_walk(self):
        # return the history of the position of the walker
        return self.walk
    


### PART 1
num_walks = 300 # perform random walk 300 times

total_positions = [] # will contain positions of all walks
total_label = [] # for creating all the labels
# creating lists to hold positions at timepoints 10,100 and 1000s for all the different walks
total_positions_10 =[]
total_positions_100 = []
total_positions_1000 = []

for i in range(num_walks): 
    walker = random_walker_equal() # creating an object of the random_walker_equal class
    walker.steps(1000) #walking 1000 steps
    walk = walker.get_walk() # returns the entire walk as a list 
    total_positions.append(walk) # nested list of all walks
    total_label.append('random walk ' + str(i+1)) # add 1 so the first walk is walk 1 and not 0
   
    # adding the 10,100 and 1000 time positions in different lists for every random walk
    total_positions_10.append(walk[10])
    total_positions_100.append(walk[100])
    total_positions_1000.append(walk[1000])

labels_for_legend = ['Time point: 10s', 'Time point: 100s', 'Time point: 1000s']

# Figure plotting

plt.rcParams['figure.figsize'] = [20, 10] # [x,y]
plt.figure(dpi=1200) #figure resolution

plt.subplot(1, 2, 1) #subplot 1 - probability density function

plt.hist(total_positions_10, histtype='step', density=True,color = 'orange')
plt.hist(total_positions_100, histtype='step', density=True, color = 'blue')
plt.hist(total_positions_1000, histtype='step', density=True, color = 'green')
plt.xlabel('Position', fontweight='bold', fontsize=15)
plt.ylabel('Probability', fontweight='bold', fontsize=15)
plt.title('Probability density function', fontweight='bold', fontsize=15)
plt.legend(labels=labels_for_legend,loc = 'upper left', fontsize=12)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

xmin, xmax, ymin, ymax = plt.axis()
scale_y = (ymax-ymin)/6
plt.text(xmin,ymax+0.1*scale_y, '(a)', fontweight='bold', fontsize=12)
# plt.tight_layout(pad=4.0)


plt.subplot(1, 2, 2) #subplot 2 - cumulative probability function

def cumulative(x, X): # x are values to evaluate F(x) and X is the random sample
    F = np.zeros(x.size)
    for i in range(x.size):        
        F[i] = (X <= x[i]).mean() 
    return F

x = np.linspace(-200, 200, 100000)
F_10 = cumulative(x, total_positions_10)
plt.plot(x, F_10,'.',color = 'orange')
F_100 = cumulative(x, total_positions_100)
plt.plot(x, F_100,'.',color = 'blue')
F_1000 = cumulative(x, total_positions_1000)
plt.plot(x, F_1000,'.',color = 'green')
plt.grid()
plt.xlabel('x', fontweight='bold', fontsize=15)
plt.ylabel('Probability', fontweight='bold', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(labels=labels_for_legend,loc = 'upper left', fontsize=12)
xmin, xmax, ymin, ymax = plt.axis()
scale_y = (ymax-ymin)/6
plt.text(xmin,ymax+0.1*scale_y, '(b)', fontweight='bold', fontsize=12)
plt.title('Cumulative probability function', fontweight='bold', fontsize=15);
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 19 13:11:40 2021

@author: preshitadave
"""

import matplotlib.pyplot as plt
import numpy as np
import random

class base_walker: 
    def __init__(self):
        pass
    def one_step(self):
        pass
    def steps(self,n=1):
        for i in range(n):
            self.one_step()
            
# inherit from base_walker to make a random_walker class that:
# stores a sequence of integer positions, and,
# increments its position by +1, 0 or -1 every step


class random_walker_biased(base_walker):
    def __init__(self):
        # define stored variables, initial data
        self.x = 0
        self.walk = [self.x]
    def one_step(self):
        rand = random.random()
        if rand > 0.5:
            self.x += 1 #moves to the right
        else:
            self.x += 0 #rests at same position
        # no weight given to moving to the left
        self.walk.append(self.x)
    def get_walk(self):
        # return the history of the position of the walker
        return self.walk
    


### PART 1
num_walks = 300 # perform random walk 300 times

total_positions = [] # will contain positions of all walks
total_label = [] # for creating all the labels
# creating lists to hold positions at timepoints 10,100 and 1000s
total_positions_10 =[]
total_positions_100 = []
total_positions_1000 = []

for i in range(num_walks): 
    walker = random_walker_biased() # creating an object of the random_walker_biased class
    walker.steps(1000)
    walk = walker.get_walk() # returns the entire walk as a list 
    total_positions.append(walk) # nested list of all walks
    total_label.append('random walk ' + str(i+1)) # add 1 so the first walk is walk 1 and not 0
   
    # adding the 10,100 and 1000 time positions in different lists for every random walk
    total_positions_10.append(walk[10])
    total_positions_100.append(walk[100])
    total_positions_1000.append(walk[1000])

labels_for_legend = ['Time point: 10s', 'Time point: 100s', 'Time point: 1000s']


# Figure plotting

plt.rcParams['figure.figsize'] = [20, 10] # [x,y]
plt.figure(dpi=1200)

plt.subplot(1, 2, 1) #subplot 1 - probability density function

plt.hist(total_positions_10, histtype='step', density=True,color = 'orange')
plt.hist(total_positions_100, histtype='step', density=True, color = 'blue')
plt.hist(total_positions_1000, histtype='step', density=True, color = 'green')
plt.xlabel('Position', fontweight='bold', fontsize=15)
plt.ylabel('Probability', fontweight='bold', fontsize=15)
plt.title('Probability density function', fontweight='bold', fontsize=15)
plt.legend(labels=labels_for_legend,loc = 'upper right', fontsize=12)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

xmin, xmax, ymin, ymax = plt.axis()
scale_y = (ymax-ymin)/6
plt.text(xmin,ymax+0.1*scale_y, '(a)', fontweight='bold', fontsize=12)
# plt.tight_layout(pad=4.0)


plt.subplot(1, 2, 2) #subplot 2 - cumulative probability function

def cumulative(x, X): # x are values to evaluate F(x) and X is the random sample
    F = np.zeros(x.size)
    for i in range(x.size):        
        F[i] = (X <= x[i]).mean()
    return F

x = np.linspace(-1000, 1000, 100000)
F_10 = cumulative(x, total_positions_10)
plt.plot(x, F_10,'.',color = 'orange')
F_100 = cumulative(x, total_positions_100)
plt.plot(x, F_100,'.',color = 'blue')
F_1000 = cumulative(x, total_positions_1000)
plt.plot(x, F_1000,'.',color = 'green')
plt.grid()
plt.xlabel('x', fontweight='bold', fontsize=15)
plt.ylabel('Probability', fontweight='bold', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(labels=labels_for_legend,loc = 'upper left', fontsize=12)
xmin, xmax, ymin, ymax = plt.axis()
scale_y = (ymax-ymin)/6
plt.text(xmin,ymax+0.1*scale_y, '(b)', fontweight='bold', fontsize=12)
plt.title('Cumulative probability function', fontweight='bold', fontsize=15);
```

# Conclusion

The Part 1 plot (a) has a more centered distribution as one can go left, right or stay at the same position; and obtaining each of these paths has an equal probability. Therefore, one cannot go very far either in the left or the right direction with the more number of steps that have been taken as it all evens out in a flatter curve. The Part 2 plot (a) differs as one can only go right or stop, and with the greater the number of steps taken, the greater would be the position to the right. Since there is equal probability to go to the right or to stop, and we take a 1000 steps, one can expect be to around the 500 step position. This is what we can also see from the plot. 

The Part 1 plot (b) plots a cumulative distribution function. We can see that they all sort of overlap very closely and the range of the higher timepoints is slightly wider because they have a greater spread as seen in Part 1 plot (a). In the Part 2 plot (b) we can see that timepoints 10 and 100s sort of lie closely to each other but the 1000s graph is near the 500 mark on the x-axis; and this can be inferred by looking at the Part 2 (a) plot.
